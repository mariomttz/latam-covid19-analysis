# ETL - EDA

```python
# Cargamos las librerías
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Declaramos la ruta de los datos
path = 'latam-covid19-analysis\\data\\'
```

---

## Argentina

```python
# Cargamos los datos
argentina = pd.read_csv(path + 'argentina\\covid_2020-2022.csv', nrows = 10000)
```

```python
# Cambiamos el formato del nombre de
# las columnas del dataset
argentina.columns = argentina.columns.str.lower()
```

### Imputing

```python
# Consultamos los datos nulos
argentina.isnull().sum()
```

### EDA

```python
# Graficamos la distribución de las edades
plt.figure(figsize = (10, 6))

sns.histplot(data = argentina, x = 'edad', bins = 30, kde = True)
plt.suptitle('Distribución de las edades en Argentina')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')

plt.show()
```

```python
# Graficamos la distribución de los sexos
plt.figure(figsize = (10, 6))

sns.countplot(data = argentina, x = 'sexo')

plt.suptitle('Distribución de los sexos en Argentina')
plt.xlabel('Sexo')
plt.ylabel('Frecuencia')

plt.show()
```

## México

```python
# Cargamos los datos de los años 
# de 2020, 2021 y 2022
mexico_2020 = pd.read_csv(path + 'mexico\\covid_2020.csv', nrows = 10000)
mexico_2021 = pd.read_csv(path + 'mexico\\covid_2021.csv', nrows = 10000)
mexico_2022 = pd.read_csv(path + 'mexico\\covid_2022.csv', nrows = 10000)
```

```python
# Cambiamos el formato del nombre de las
# columnas de los datasets
mexico_2020.columns = mexico_2020.columns.str.lower()
mexico_2021.columns = mexico_2021.columns.str.lower()
mexico_2022.columns = mexico_2022.columns.str.lower()
```

### Imputing

```python
# Consultamos los datos nulos del primer dataset
mexico_2020.isnull().sum()
```

```python
# Consultamos los datos nulos del segundo dataset
mexico_2021.isnull().sum()
```

```python
# Consultamos los datos nulos del tercer dataset
mexico_2022.isnull().sum()
```

### EDA

```python
# Graficamos la distribución de las edades
# de los años 2020, 2021 y 2022
fig, axes = plt.subplots(1, 3, figsize = (15, 5))
fig.suptitle('Distribución de las edades en México')

sns.histplot(data = mexico_2020, x = 'edad', ax = axes[0], kde = True)
axes[0].set_title('Año 2020')
axes[0].set_xlabel('Edad')
axes[0].set_ylabel('Frecuencia')

sns.histplot(data = mexico_2021, x = 'edad', ax = axes[1], kde = True)
axes[1].set_title('Año 2021')
axes[1].set_xlabel('Edad')
axes[1].set_ylabel('Frecuencia')

sns.histplot(data = mexico_2022, x = 'edad', ax = axes[2], kde = True)
axes[2].set_title('Año 022')
axes[2].set_xlabel('Edad')
axes[2].set_ylabel('Frecuencia')

plt.show()
```

```python
# Graficamos la distribución de los sexos
# de los años 2020, 2021 y 2022
fig, axes = plt.subplots(1, 3, figsize = (15, 5))
fig.suptitle('Distribución de los sexos en México')

sns.countplot(data = mexico_2020, x = 'sexo', ax = axes[0])
axes[0].set_title('Año 2020')
axes[0].set_xlabel('Sexo')
axes[0].set_ylabel('Frecuencia')

sns.countplot(data = mexico_2021, x = 'sexo', ax = axes[1])
axes[1].set_title('Año 2021')
axes[1].set_xlabel('Sexo')
axes[1].set_ylabel('Frecuencia')

sns.countplot(data = mexico_2022, x = 'sexo', ax = axes[2])
axes[2].set_title('Año 2022')
axes[2].set_xlabel('Sexo')
axes[2].set_ylabel('Frecuencia')

plt.show()
```

## Perú

```python
# Cargamos el primer y segundo conjunto de datos
peru_positivos  = pd.read_csv(path + 'peru\\positivos_covid_2020-2022.csv',  sep=';', nrows = 10000)
peru_fallecidos = pd.read_csv(path + 'peru\\fallecidos_covid_2020-2022.csv', sep=';', nrows = 10000)
```

```python
# Cambiamos el formato del nombre de las
# columnas de los datasets
peru_positivos.columns = peru_positivos.columns.str.lower()
peru_fallecidos.columns = peru_fallecidos.columns.str.lower()
```

### Imputing

```python
# Contamos los valores nulos del primer
# conjunto de datos
peru_positivos.isnull().sum()
```

```python
# Contamos los valores nulos del segundo
# conjunto de datos
peru_fallecidos.isnull().sum()
```

### EDA

```python
# Graficamos la distribución de las edades
# de los positivos y fallecidos
fig, axes = plt.subplots(1, 2, figsize = (15, 5))
fig.suptitle('Distribución de las edades en Perú')

sns.histplot(data = peru_positivos, x = 'edad', ax = axes[0], kde = True)
axes[0].set_title('Positivos')
axes[0].set_xlabel('Edad')
axes[0].set_ylabel('Frecuencia')

sns.histplot(data = peru_fallecidos, x = 'edad_declarada', ax = axes[1], kde = True)
axes[1].set_title('Fallecidos')
axes[1].set_xlabel('Edad')
axes[1].set_ylabel('Frecuencia')

plt.show()
```

```python
# Graficamos la distribución de los sexos
# de los positivos y fallecidos
fig, axes = plt.subplots(1, 2, figsize = (15, 5))
fig.suptitle('Distribución de los sexos en Perú')

sns.countplot(data = peru_positivos, x = 'sexo', ax = axes[0])
axes[0].set_title('Positivos')
axes[0].set_xlabel('Sexo')
axes[0].set_ylabel('Frecuencia')

sns.countplot(data = peru_fallecidos, x = 'sexo', ax = axes[1])
axes[1].set_title('Fallecidos')
axes[1].set_xlabel('Sexo')
axes[1].set_ylabel('Frecuencia')

plt.show()
```
